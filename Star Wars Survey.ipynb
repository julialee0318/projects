# Star Wars Survey

The objective of this project is to analyze the survey data about Star Wars movies to compare which is the highest ranked movie and the highest viewed movie in the series. 

Data set used in this project can be found in the following link.
https://github.com/fivethirtyeight/data/tree/master/star-wars-survey

# Read the Data

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

# Explore the Data

```python
star_wars.head(10)
```

```python
star_wars.columns
```

# Clean and Map Yes/No Columns

The following two columns currently store string data types:
- `Have you seen any of the 6 films in the Star Wars franchise?`
- `Do you consider yourself to be a fan of the Star Wars film franchise?`
We are going to convert these two columns from string to boolean data types for easy analysis.

```python
# Check the unique values of the columns
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False))
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False))
```

```python
# Convert 'Have you seen any of the 6 films in the Star Wars franchise?' to Boolean
yes_no = {
    "Yes": True,
    "No": False
}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

# Clean and Map Checkbox Columns

There are six columns that represent a single checkbox question:
- `Which of the following Star Wars films have you seen? Please select all that apply.`: whether or not the respondent saw Star Wars: Episode I The Phantom Menace
- `Unnamed: 4`: whether or not the respondent saw Star Wars: Episode II Attack of the Clones
- `Unnamed: 5`: whether or not the respondent saw Star Wars: Episode III Revenge of the Sith
- `Unnamed: 6`: whether or not the respondent saw Star Wars: Episode IV A New Hope
- `Unnamed: 7`: whether or not the respondent saw Star Wars: Episode V The Empire Strikes Back
- `Unnamed: 8`: Star Wars: Episode VI Return of the Jedi

We can see that if the respondent saw the movie, the value of the corresponding column would be the name of the movie. On the contrary, if the repsondant did not see the movie, the value will be `NaN`.

We are going to convert these columns into Boolean data type as well.

```python
# Check the values of these columns
seen_cols = star_wars.columns[3:9]
for col in seen_cols:
    print(star_wars[col].value_counts(dropna=False))
```

```python
# Convert the columns into Boolean data type
import numpy as np

seen_movie = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}

for col in seen_cols:
    star_wars[col] = star_wars[col].map(seen_movie)
```

```python
# Rename these columns with the movie names
star_wars = star_wars.rename(columns={
        'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
        'Unnamed: 4': 'seen_2',
        'Unnamed: 5': 'seen_3',
        'Unnamed: 6': 'seen_4',
        'Unnamed: 7': 'seen_5',
        'Unnamed: 8': 'seen_6'
})

star_wars.head()
```

# Clean the Ranking Columns

There are six columns in the survey that ask the respondent to rank the Star Wars movies from least to most favorite. These columns contain values of 1, 2, 3, 4, 5, 6 or `NaN`.

These six columns are:
- `Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.` - How much the respondent liked Star Wars: Episode I The Phantom Menace
- `Unnamed: 10` — How much the respondent liked Star Wars: Episode II Attack of the Clones
- `Unnamed: 11` — How much the respondent liked Star Wars: Episode III Revenge of the Sith
- `Unnamed: 12` — How much the respondent liked Star Wars: Episode IV A New Hope
- `Unnamed: 13` — How much the respondent liked Star Wars: Episode V The Empire Strikes Back
- `Unnamed: 14` — How much the respondent liked Star Wars: Episode VI Return of the Jedi

We are going to convert the values from string to float, which allow us to calculate the average 'score' of popularity rated by the respondants in our analysis.

```python
# Convert the data type from string to float
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Rename the columns
# Rename these columns with the movie names
star_wars = star_wars.rename(columns={
        'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
        'Unnamed: 10': 'ranking_2',
        'Unnamed: 11': 'ranking_3',
        'Unnamed: 12': 'ranking_4',
        'Unnamed: 13': 'ranking_5',
        'Unnamed: 14': 'ranking_6'
})

star_wars.head()
```

# The Highest Ranked Movie

After cleaning the ranking columns, we are going to find the highest ranked movie by calculating the average 'score' given in each movie.

```python
# Compute the mean of each column
ranking_mean = star_wars[star_wars.columns[9:15]].agg(np.mean)
print(ranking_mean)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects = ax.bar(ind, ranking_mean, width, color='r')

# add some text for labels, title and axes ticks
ax.set_ylabel('Average rank')
ax.set_title('Average Rank of Movies')
ax.set_xticks(ind)
ax.set_xticklabels(('Episode I', 'Episode II', 'Episode III', 'Episode IV', 'Episode V', 'Episode VI'))

plt.show()
```

Since 1 represents the most favorite movie and 6 represent the least favorite movie, movie that receives lower scores mean that they are ranked higher by the respondent.

From the chart above, we can see that the `Star Wars: Episode V The Empire Strikes Back` received the lowest score, thus is ranked the highest among the respondents.

We can also see that `Star Wars: Episode III Revenge of the Sith` is the least favorite among the respondents.

# The Most Viewed Movie

We are going to find the most seen movie by the respondents in this section.

```python
# Compute the sum of all the 'seen' columns
seen_sum = star_wars[star_wars.columns[3:9]].agg(np.sum)
print(seen_sum)
```

```python
N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects = ax.bar(ind, seen_sum, width, color='r')

# add some text for labels, title and axes ticks
ax.set_ylabel('Total views')
ax.set_title('Total Views of Movies')
ax.set_xticks(ind)
plt.ylim(0, 850)
ax.set_xticklabels(('Episode I', 'Episode II', 'Episode III', 'Episode IV', 'Episode V', 'Episode VI'))

def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects)

plt.show()
```

From the chart above, we observe that the most watched movie is `Star Wars: Episode V The Empire Strikes Back`, which corresponds to the most favorite movie ranked by respondents.

The same applies to the least watched movie, in this case `Star Wars: Episode III Revenge of the Sith`, for the least favorite movie.

It would be interesting to analyze the factors, e.g. plot, production, advertisment, etc. that lead to such results in another study.

# Star Wars Fans VS Non Star Wars Fans

Another interesting angle to analyze the data set is to look at the data by whether the respondent is a Star Wars fans or not. We can identify this by the following column:

- `Do you consider yourself to be a fan of the Star Wars film franchise?`

We are going to split the data into these two segments for analysis in this section.

```python
# Rename the two columns to shorten the column names
star_wars = star_wars.rename(columns={
        'Do you consider yourself to be a fan of the Star Wars film franchise?': 'star_wars_fans',
})
```

```python
# Split star_wars_fans column
sw_fans = star_wars[star_wars['star_wars_fans'] == True]
non_sw_fans = star_wars[star_wars['star_wars_fans'] == False]
```

```python
# Ranking mean of two segments
ranking_mean_sw = sw_fans[star_wars.columns[9:15]].agg(np.mean)
print(ranking_mean_sw)

ranking_mean_nsw = non_sw_fans[star_wars.columns[9:15]].agg(np.mean)
print(ranking_mean_nsw)
```

```python
# Highest ranked movies by these two segments
N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects_sw = ax.bar(ind, ranking_mean_sw, width, color='r')
rects_nsw = ax.bar(ind + width, ranking_mean_nsw, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Average rank')
ax.set_title('Average Rank By Movie')
ax.set_xticks(ind + width / 2)
plt.ylim(0,5)
ax.set_xticklabels(('Episode I', 'Episode II', 'Episode III', 'Episode IV', 'Episode V', 'Episode VI'))

ax.legend((rects_sw[0], rects_nsw[0]), ('Star Wars Fans', 'Non Star Wars Fans'))

plt.show()
```

From the chart above, we see that the highest ranked movie by Star Wars fans is `Star Wars: Episode V The Empire Strikes Back`.

Whereas for non Star Wars, they rated both `Star Wars: Episode V The Empire Strikes Back` and `Star Wars: Episode I The Phantom Menace` as their favorite.

```python
# View sum of two segments
seen_sw = sw_fans[star_wars.columns[3:9]].agg(np.sum)
print(seen_sw)

seen_nsw = non_sw_fans[star_wars.columns[3:9]].agg(np.sum)
print(seen_sw)
```

```python
N = 6
ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects_sw = ax.bar(ind, seen_sw, width, color='r')
rects_nsw = ax.bar(ind + width, seen_nsw, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Total views')
ax.set_title('Total Views of Movies')
ax.set_xticks(ind + width / 2)
plt.ylim(0,650)
ax.set_xticklabels(('Episode I', 'Episode II', 'Episode III', 'Episode IV', 'Episode V', 'Episode VI'))

ax.legend((rects_sw[0], rects_nsw[0]), ('Star Wars Fans', 'Non Star Wars Fans'))

autolabel(rects_sw)
autolabel(rects_nsw)

plt.show()
```

We see that there is a remarkable difference in the total views of the movies between Star Wars fans and non Star Wars fans. On average, 60% of the views come from fans.

`Star Wars: Episode V The Empire Strikes Back` is the top viewed movie among all for both fans and non fans, while `Star Wars: Episode III Revenge of the Sith` is the least viewed among both segments.

# Findings

From this prelimiary analysis, it suggests that the favorite movie in the Star Wars movie series is `Star Wars: Episode V The Empire Strikes Back` which is also the most viewed movie among all.
