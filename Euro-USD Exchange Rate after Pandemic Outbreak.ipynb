# Euro-USD Exchange Rate after Pandemic Period Outbreak Visualization

The objective of this project is to create visualization of the changes in Euro-USD exchange rate after 2020's coronavirus pandemic outbreak.

The dataset we'll use describes Euro daily exchange rates between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union. The data set can be found in this [link](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020).

We will be focusing on the period of 2016-2020 to look at the effect of how the euro-dollar rate has changed during the coronavirus pandemic.

## Result Summary

We created a line graph with moving average of daily exchange rate. The moving window is 30 days. The graph is presented at the end of the project.

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

## Clean the data

### Rename columns: `US dollar` and `Period\\Unit`

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

### Isolate relevant columns

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

In this section, we have isolated the columns of `Time` and `US_dollar` since this is the exchange rate we want to focus on.

We have also cleaned the data for easy analysis later by renaming these columns and removing unuseful information.

## Explore data: Evolution of euro-dollar exchange rate

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8,3))
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

plt.figure(figsize=(8,3))
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.title('Euro-USD Exchange Rate Between 2000-2020', size=12)
plt.ylabel('Rate')
plt.show()
```

In this section, we have created a graph of 30 days moving average of daily exchange rate of Euro-USD.

With this given information, we want to focus on visualizing how the euro-dollar rate has changed during the coronavirus pandemic. We will use the 2020 data and the 2016-2019 data as a baseline.

## Result: Euro-dollar rate during coronavirus pandemic

```python
pandemic_euro = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2016) & 
    (euro_to_dollar['Time'].dt.year <= 2021)]

pandemic_euro_2020 = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2020) & 
    (euro_to_dollar['Time'].dt.year <= 2021)]
```

```python
pandemic_euro_2020['rolling_mean'].min()
```

```python
# adding the style
import matplotlib.style as style
style.use('fivethirtyeight')

# adding the plot
fig, ax = plt.subplots(figsize=(13,6))
ax.plot(pandemic_euro['Time'],
        pandemic_euro['rolling_mean'],
        linewidth=1, color='#8ec127')

# adding the highlight part for pandemic period
ax.plot(pandemic_euro_2020['Time'],
        pandemic_euro_2020['rolling_mean'],
        linewidth=3, color='#ff0000')

# setting separate x, y labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 735964
for year in  ['2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, 1.03, year, alpha=0.5, fontsize=15)
    x += 365
    
y = 1.05
for rate in ['1.05', '1.1', '1.15', '1.2','1.25']:
    ax.text(735750, y, rate, alpha=0.5, fontsize=15)
    y += 0.05

# highlight the period of pandemic
ax.axvspan(xmin=737608, xmax=737425, ymin=0.03,
           alpha=0.3, color='grey')

# adding a title and subtitle
ax.text(735750, 1.28, 'EURO-USD rate troughed at 1.08 and recovered after pandemic outbreak', size=20, weight='bold')
ax.text(735750, 1.265, 'EURO-USD exchange rates between 2016 and 2020', size=18)

# adding the data source
ax.text(735750, 1, '©DATAQUEST' + ' '*110 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=15)

plt.show()
```
